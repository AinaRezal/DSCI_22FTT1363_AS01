```python
import pandas as pd
import numpy as np

df = pd.read_csv('earthquake.csv')
df.head()
```

```python
df.isnull().sum()
```

DROPPING COLUMNS

```python
# Dropping 3 columns

df.drop(['country', 'continent', 'location'], axis='columns', inplace=True)
df.head()
```

FILLING IN MISSING VALUES

```python
# Filling in missing values of column 'alert' with 'unknown'

df['alert'] = df['alert'].fillna('unknown')
df.isnull().sum()
```

DUPLICATES

```python
# Checking for duplicated rows

df.duplicated().sum()
```

SAVING INTO NEW FILE

```python
# Creating a new csv file

df.to_csv('clean_earthquake.csv', index=False)
```
