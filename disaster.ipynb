```python
import pandas as pd
import numpy as np

df = pd.read_csv('natural_disasters.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Reconstruction costs from earthquakes',
         'Insured damages against earthquakes',
         'Total economic damages from earthquakes',
         'Total economic damages from earthquakes as a share of GDP'],
          inplace=True)
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```

```python
# Creating a new CSV file

df.to_csv('clean_disaster.csv', index=False)
```
